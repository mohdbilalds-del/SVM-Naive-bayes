#Theoretical

**Q1. What is a Support Vector Machine (SVM)?**
- A Support Vector Machine (SVM) is a supervised machine learning algorithm primarily used for classification tasks, though it can also handle regression. The main idea is to find the optimal hyperplane that separates data points of different classes with the maximum margin. By maximizing this margin, SVM improves generalization and reduces the risk of misclassification on unseen data. It is particularly effective in high-dimensional spaces and can be extended to non-linear problems using kernel functions.


**Q2. What is the difference between Hard Margin and Soft Margin SVM?**
- Hard Margin SVM assumes that the training data is perfectly linearly separable, meaning no misclassification is allowed. It finds a hyperplane that separates classes with maximum margin but fails when data contains noise or overlaps. Soft Margin SVM, on the other hand, introduces slack variables that allow some misclassifications. This flexibility helps the model generalize better to real-world data, where perfect separation is rare. The trade-off between margin size and misclassification is controlled by the parameter C.


**Q3. What is the mathematical intuition behind SVM?**
- The mathematical intuition behind SVM lies in maximizing the margin between classes. The optimization problem minimizes the norm of the weight vector while ensuring that all data points are correctly classified with a margin of at least one.

  Formally, it solves:
  $[\min \frac{1}{2}||w||^2 \quad \text{subject to } y_i(w \cdot x_i + b) \geq 1]$
  
  This ensures the hyperplane is positioned to maximize separation. By maximizing margin, SVM reduces overfitting and improves robustness to unseen data.


**Q4. What is the role of Lagrange Multipliers in SVM?**
- Lagrange Multipliers are used in SVM to transform the constrained optimization problem into its dual form. This dual representation allows the algorithm to focus only on support vectors rather than all data points, making computation more efficient. Moreover, the dual form enables the use of kernel functions, which project data into higher-dimensional spaces without explicitly computing transformations. Thus, Lagrange Multipliers are essential for handling non-linear decision boundaries and making SVM practical for complex datasets.


**Q5. What are Support Vectors in SVM?**
- Support Vectors are the data points that lie closest to the decision boundary (hyperplane). They are critical because they define the margin and directly influence the position of the hyperplane. Removing non-support vectors does not affect the decision boundary, but removing support vectors changes it significantly. In essence, support vectors are the most informative points in the dataset, and the SVM algorithm relies on them to achieve optimal classification. This property makes SVM computationally efficient


**Q6. What is a Support Vector Classifier (SVC)?**
- A Support Vector Classifier (SVC) is the classification implementation of SVM. It works by finding the optimal hyperplane that separates data points into distinct classes. Unlike simple linear classifiers, SVC can handle non-linear boundaries using kernel functions, making it versatile for complex datasets. It is particularly useful in high-dimensional spaces, such as text classification or image recognition, where traditional algorithms struggle. SVC’s strength lies in its ability to maximize margin and generalize well.


**Q7. What is a Support Vector Regressor (SVR)?**
- Support Vector Regressor (SVR) is an extension of SVM designed for regression tasks. Instead of finding a hyperplane to separate classes, SVR fits a function within a margin of tolerance around the data points. The goal is to minimize prediction error while keeping the model simple. SVR uses kernels to capture non-linear relationships, making it effective for complex regression problems. It balances accuracy and generalization, ensuring predictions are robust even with noisy or high-dimensional data.


**Q8. What is the Kernel Trick in SVM?**
- The Kernel Trick is a mathematical technique that allows SVM to handle non-linear data by implicitly mapping it into higher-dimensional spaces. Instead of explicitly computing transformations, kernel functions (like polynomial or RBF) calculate inner products in the transformed space. This makes it computationally efficient and enables SVM to find complex decision boundaries. The Kernel Trick is the reason SVM can solve problems where linear separation is impossible, making it a powerful tool for diverse applications.


**Q9. Compare Linear Kernel, Polynomial Kernel, and RBF Kernel.**
- **Linear Kernel:** Best for linearly separable data; simple and fast.
- **Polynomial Kernel:** Captures more complex relationships by introducing polynomial decision boundaries; useful when interactions between features matter.
- **RBF Kernel:** Highly flexible, handles non-linear data by creating localized decision boundaries; widely used in practice.
Each kernel has trade-offs: linear is efficient, polynomial can overfit with high degrees, and RBF requires careful tuning of parameters like gamma. Choosing the right kernel depends on dataset complexity.


**Q10. What is the effect of the C parameter in SVM?**
- The C parameter controls the trade-off between maximizing margin and minimizing classification errors. A high C value forces the model to classify training examples correctly, even if it means a smaller margin, which can lead to overfitting. A low C value allows more misclassifications but results in a wider margin, improving generalization. Essentially, C determines how strictly the model penalizes misclassified points, making it a crucial hyperparameter for balancing bias and variance.


**Q11. What is the role of the Gamma parameter in RBF Kernel SVM?**
- Gamma defines how far the influence of a single training example reaches. A high gamma means each point has a very localized effect, creating complex, tightly curved decision boundaries that may overfit. A low gamma spreads influence more broadly, resulting in smoother boundaries but potentially underfitting. In practice, gamma controls the flexibility of the RBF kernel. Proper tuning of gamma, along with C, is essential to achieve a balance between accuracy and generalization.


**Q12. What is the Naïve Bayes classifier, and why is it called "Naïve"?**
- Naïve Bayes is a probabilistic classifier based on Bayes’ theorem. It predicts class membership by calculating the probability of a class given the input features. The “naïve” assumption is that all features are conditionally independent of each other given the class label. While this assumption is rarely true in practice, the model still performs surprisingly well, especially in text classification tasks, because independence simplifies computation and reduces the risk of overfitting.


**Q13. What is Bayes’ Theorem?**
- Bayes’ theorem provides a way to calculate conditional probabilities.

  It states: $P(A|B)=\frac{P(B|A)\cdot P(A)}{P(B)}$

  Here, $P(A|B)$ is the probability of event A given event B. It combines prior probability $P(A)$ with likelihood $P(B|A)$ to update beliefs after observing evidence. In machine learning, Bayes’ theorem underpins Naïve Bayes, allowing the model to estimate class probabilities based on observed features and prior knowledge.


**Q14. Explain the differences between Gaussian Naïve Bayes, Multinomial Naïve Bayes, and Bernoulli Naïve Bayes.**
- **Gaussian Naïve Bayes:** Assumes continuous features follow a normal distribution. Useful for datasets with continuous variables like sensor readings.
- **Multinomial Naïve Bayes:** Works with discrete counts, such as word frequencies in text classification.
- **Bernoulli Naïve Bayes:** Designed for binary features, such as presence or absence of a word.
Each variant adapts the independence assumption to different data types, making Naïve Bayes versatile across varied domains.


**Q15. When should you use Gaussian Naïve Bayes over other variants?**
- Gaussian Naïve Bayes is best suited when features are continuous and approximately normally distributed. For example, in medical datasets with measurements like blood pressure or temperature, Gaussian assumptions hold reasonably well. Unlike Multinomial or Bernoulli, which are tailored for text or categorical data, Gaussian handles continuous variables naturally. It is also computationally efficient, making it a good choice when speed and simplicity are important, even if independence assumptions are not perfectly met.


**Q16. What are the key assumptions made by Naïve Bayes?**
- Naïve Bayes assumes that all features are conditionally independent given the class label. This means the presence or absence of one feature does not affect another, which simplifies probability calculations. It also assumes that each feature contributes equally to the classification decision. While these assumptions are rarely true in real-world data, they make the model computationally efficient and surprisingly effective, especially in domains like text classification where independence approximations work reasonably well.


**Q17. What are the advantages and disadvantages of Naïve Bayes?**
- **Advantages:** Naïve Bayes is simple, fast, and efficient, even with small datasets. It performs well in high-dimensional spaces like text classification and requires little training data.
- **Disadvantages:** The independence assumption is often unrealistic, leading to reduced accuracy when features are correlated. It also struggles with zero-frequency problems, where unseen features result in zero probability. Despite these limitations, Naïve Bayes remains popular due to its robustness and ease of implementation.


**Q18. Why is Naïve Bayes a good choice for text classification?**
- Naïve Bayes is highly effective for text classification tasks such as spam detection or sentiment analysis. Text data often involves thousands of features (words), and Naïve Bayes handles this high dimensionality well. The independence assumption works reasonably because word occurrences can be treated as independent signals. Additionally, it is computationally efficient, requiring minimal training data, and produces accurate results quickly. Its simplicity and speed make it a strong baseline for text-related machine learning problems.


**Q19. Compare SVM and Naïve Bayes for classification tasks.**
- SVM and Naïve Bayes differ in approach and performance. SVM is a margin-based classifier that excels in complex, high-dimensional datasets, offering strong accuracy but requiring more computational resources. Naïve Bayes is probabilistic, faster to train, and works well with text data, but its independence assumption can limit accuracy. SVM is better for non-linear problems with kernel tricks, while Naïve Bayes is preferred for quick, interpretable solutions. The choice depends on dataset complexity and speed requirements.


**Q20. How does Laplace Smoothing help in Naïve Bayes?**
- Laplace Smoothing addresses the zero-frequency problem in Naïve Bayes. Without smoothing, if a feature never appears in the training data for a class, its probability becomes zero, which can incorrectly eliminate that class during prediction. Laplace Smoothing adds a small constant (usually 1) to all frequency counts, ensuring no probability is zero. This adjustment improves robustness, especially in text classification, where unseen words are common, and helps the model generalize better to new data.

```python
#Q21. Write a Python program to train an SVM Classifier on the Iris dataset and evaluate accuracy.

from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Load dataset
iris = datasets.load_iris()
X, y = iris.data, iris.target

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train SVM
model = SVC(kernel='linear')
model.fit(X_train, y_train)

# Predict and evaluate
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
#Q22. Write a Python program to train two SVM classifiers with Linear and
#RBF kernels on the Wine dataset, then compare their accuracies.

from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

wine = load_wine()
X, y = wine.data, wine.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Linear kernel
linear_svm = SVC(kernel='linear')
linear_svm.fit(X_train, y_train)
linear_acc = accuracy_score(y_test, linear_svm.predict(X_test))

# RBF kernel
rbf_svm = SVC(kernel='rbf', gamma='scale')
rbf_svm.fit(X_train, y_train)
rbf_acc = accuracy_score(y_test, rbf_svm.predict(X_test))

print("Linear Kernel Accuracy:", linear_acc)
print("RBF Kernel Accuracy:", rbf_acc)
```

```python
#Q24. Write a python program to train an SVM Classifier with a Polynomial Kernel
#and visualize the decision boundary.

import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_classification
from sklearn.svm import SVC

# Generate synthetic dataset
X, y = make_classification(n_features=2, n_redundant=0, n_informative=2, n_clusters_per_class=1, n_samples=200)

# Train polynomial kernel SVM
poly_svm = SVC(kernel='poly', degree=3)
poly_svm.fit(X, y)

# Plot decision boundary
plt.scatter(X[:,0], X[:,1], c=y, cmap='coolwarm')
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 100),
                     np.linspace(ylim[0], ylim[1], 100))
Z = poly_svm.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')
plt.title("Polynomial Kernel SVM Decision Boundary")
plt.show()
```

```python
#Q27. Write a Python program to train an SVM Classifier with different C values and compare decision boundaries visually.

import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_classification
from sklearn.svm import SVC

X, y = make_classification(n_features=2, n_redundant=0, n_informative=2, n_clusters_per_class=1, n_samples=200)

for C in [0.1, 1, 10]:
    model = SVC(kernel='linear', C=C)
    model.fit(X, y)
    plt.scatter(X[:,0], X[:,1], c=y, cmap='coolwarm')
    ax = plt.gca()
    xx, yy = np.meshgrid(np.linspace(ax.get_xlim()[0], ax.get_xlim()[1], 100),
                         np.linspace(ax.get_ylim()[0], ax.get_ylim()[1], 100))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
    plt.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')
    plt.title(f"SVM Decision Boundary (C={C})")
    plt.show()
```

```python
#Q29. Write a Python program to apply feature scaling before training an SVM model and compare results with unscaled data.

from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_wine
from sklearn.metrics import accuracy_score

wine = load_wine()
X, y = wine.data, wine.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Without scaling
model_unscaled = SVC(kernel='rbf')
model_unscaled.fit(X_train, y_train)
acc_unscaled = accuracy_score(y_test, model_unscaled.predict(X_test))

# With scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
model_scaled = SVC(kernel='rbf')
model_scaled.fit(X_train_scaled, y_train)
acc_scaled = accuracy_score(y_test, model_scaled.predict(X_test))

print("Unscaled Accuracy:", acc_unscaled)
print("Scaled Accuracy:", acc_scaled)
```

```python
#Q31. Write a Python program to train an SVM Classifier and use GridSearchCV to tune the hyperparameters (C, gamma, kernel).

from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC

# Load dataset
wine = load_wine()
X, y = wine.data, wine.target

# Split into train/test first
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Define parameter grid
param_grid = {
    'C': [0.1, 1, 10],
    'gamma': ['scale', 'auto'],
    'kernel': ['linear', 'rbf']
}

# Grid search
grid = GridSearchCV(SVC(), param_grid, cv=3)
grid.fit(X_train_scaled, y_train)

print("Best Parameters:", grid.best_params_)
print("Best Accuracy:", grid.best_score_)
```

```python
#Q35. Write a Python program to perform feature selection before training a Naïve Bayes classifier and compare results.

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

# Load dataset
cancer = load_breast_cancer()
X, y = cancer.data, cancer.target

# Split into train/test
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# -------- Without Feature Selection --------
nb = GaussianNB()
nb.fit(X_train, y_train)
y_pred = nb.predict(X_test)
acc_no_fs = accuracy_score(y_test, y_pred)

# -------- With Feature Selection --------
selector = SelectKBest(chi2, k=10)   # select top 10 features
X_train_selected = selector.fit_transform(X_train, y_train)
X_test_selected = selector.transform(X_test)

nb_fs = GaussianNB()
nb_fs.fit(X_train_selected, y_train)
y_pred_fs = nb_fs.predict(X_test_selected)
acc_fs = accuracy_score(y_test, y_pred_fs)

print("Accuracy without Feature Selection:", acc_no_fs)
print("Accuracy with Feature Selection:", acc_fs)
```

```python
#Q36. Write a Python program to train an SVM Classifier using One-vs-Rest (OvR) and
#One-vs-One (OvO) strategies on the Wine dataset and compare their accuracy.

from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.multiclass import OneVsRestClassifier, OneVsOneClassifier
from sklearn.metrics import accuracy_score

wine = load_wine()
X, y = wine.data, wine.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

ovr = OneVsRestClassifier(SVC(kernel='linear'))
ovo = OneVsOneClassifier(SVC(kernel='linear'))

ovr.fit(X_train, y_train)
ovo.fit(X_train, y_train)

print("OvR Accuracy:", accuracy_score(y_test, ovr.predict(X_test)))
print("OvO Accuracy:", accuracy_score(y_test, ovo.predict(X_test)))
```

```python
#Q46. Write a Python program to train an SVM Classifier and visualize the Precision-Recall Curve.

from sklearn.metrics import precision_recall_curve
import matplotlib.pyplot as plt

model = SVC(kernel='linear', probability=True)
model.fit(X_train, y_train)
y_prob = model.predict_proba(X_test)[:,1]

precision, recall, thresholds = precision_recall_curve(y_test, y_prob)
plt.plot(recall, precision, marker='.')
plt.title("Precision-Recall Curve")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.show()
```
